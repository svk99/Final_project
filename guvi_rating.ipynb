```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df=pd.read_csv('/content/3.1-data-sheet-guvi-courses.csv')
```

```python
df
```

```python
df.drop(['course_id','course_title','url','published_timestamp'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df=df.dropna(axis=0)
```

```python
df=df.drop_duplicates()
```

```python
df.describe()
```

```python
df.shape
```

```python
outlier_cols=['num_subscribers','num_reviews','num_lectures','content_duration']
for c in outlier_cols:
  Q1=df[c].quantile(0.25)
  Q3=df[c].quantile(0.75)
  IQR=Q3-Q1
  LC=Q1-(IQR*1.5)
  UC=Q3+(IQR*1.5)
  df=df[(df[c]>LC) & (df[c]<UC)]
```

```python
df.shape
```

```python
df.describe()
```

```python
df['level'].unique()
```

```python
le=LabelEncoder()
df['level']=le.fit_transform(df['level'])
df=pd.get_dummies(df,columns=['subject'])
```

```python
df['level'].unique()
```

```python
df.reset_index(inplace=True)
df.drop(['index'],axis=1,inplace=True)
```

```python
df
```

```python
X=df.drop(['Rating'],axis=1)
y=df['Rating']
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
model=RandomForestRegressor()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
mse=mean_squared_error(y_pred,y_test)
r2=r2_score(y_test,y_pred)
print(f'Mean Squared Error-> {mse}')
print(f'R2 score-> {r2}')
```
